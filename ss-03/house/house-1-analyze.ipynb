```python
pd.set_option("display.max_rows", None)
data_types = pd.DataFrame(columns=["name", "size", "dtype", "variation", "types"])
columns = train.columns.drop("Id")

for idx in columns:
  column = train[idx]
  size = len(column.dropna())
  dtype = column.dtype
  types = sorted(column.dropna().unique())
  row = { "name": idx, "size": size, "dtype": dtype, "variation": len(types), "types": types }
  data_types = data_types.append(row, ignore_index=True)

data_types
```

```python
data_types.to_csv("./data-types.csv", index_label = ["No."])
```

```python
corr = train.corr()
# corr[(corr > 0.4) | (corr < -0.4)]
sale_corr = corr[["SalePrice"]]
sale_corr.query("SalePrice < -0.4 | 0.4 < SalePrice")
```

```python
high_corr_columns = sale_corr.query("SalePrice < -0.4 | 0.4 < SalePrice").index
high_corr_columns
```

```python
# pd.set_option("display.max_rows", None)
data_types = pd.DataFrame(columns=["name", "size", "dtype", "variation", "types"])

for idx in high_corr_columns:
  column = train[idx]
  size = len(column.dropna())
  dtype = column.dtype
  types = sorted(column.dropna().unique())
  row = { "name": idx, "size": size, "dtype": dtype, "variation": len(types), "types": types }
  data_types = data_types.append(row, ignore_index=True)

data_types
```

```python
data_types.to_csv("./data-types-high-corr.csv", index_label = ["No"])
```
